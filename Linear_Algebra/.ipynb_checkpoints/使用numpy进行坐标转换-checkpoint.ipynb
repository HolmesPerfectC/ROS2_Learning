```python
import math
```

```python
def rz(theta):
    return np.asarray([math.cos(theta), -math.sin(theta), 0, math.sin(theta), math.cos(theta), 0, 0, 0, 1]).reshape(3, 3)
```

```python
rz(math.pi / 4)
```

```python
def rx(theta):
    return np.asarray([1, 0, 0, 0, math.cos(theta), -math.sin(theta), 0, math.sin(theta), math.cos(theta)]).reshape(3, 3)
```

```python
def ry(theta):
    return np.asarray([math.cos(theta), 0, math.sin(theta), 0, 1, 0, -math.sin(theta), 0, math.cos(theta)]).reshape(3, 3)
```

```python
R_BC = rx(math.radians(180))
```

```python
R_BC
```

```python
P_CP = np.asarray([2, 1, 2])
P_BC = np.asarray([0, 0, 3])
P_BC = np.add(np.dot(R_BC, P_CP), P_BC)
```
